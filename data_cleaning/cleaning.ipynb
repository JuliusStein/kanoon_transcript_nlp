```python
caseFrame = pd.read_json('../caseData_all.json')
```

```python
caseFrame = caseFrame.replace('\n','', regex=True)
caseFrame.head()
```

```python
resultCode = []
dressCode = []
for index, row in caseFrame.iterrows():
    if('neither' in row['resolution'].lower()):
        resultCode.append("U")
    elif('acquit' in row['resolution'].lower()):
        resultCode.append("A")
    elif('convict' in row['resolution'].lower()):
        resultCode.append("C")
    elif('dismiss' in row['resolution'].lower()):
        resultCode.append("A")
    else:
        resultCode.append("U")
    
    if(row['dressMentioned'].strip().lower() in ['yes', 'yes.']):
        dressCode.append("Y")
    elif(row['dressMentioned'].strip().lower() in ['no', 'no.']):
        dressCode.append("N")
    elif(row['dressMentioned'].strip().lower() in ['maybe', 'maybe.']):
        dressCode.append("M")
    elif('not mentioned' in row['dressMentioned'].lower()):
        dressCode.append("N")
    elif('mentioned' in row['dressMentioned'].lower()):
        dressCode.append("Y")
    elif('came up' in row['dressMentioned'].lower()):
        dressCode.append("Y")
    elif('not come up' in row['dressMentioned'].lower()):
        dressCode.append("N")
    elif('come up' in row['dressMentioned'].lower()):
        dressCode.append("Y")
    else:
        dressCode.append("M")

caseFrame['resultCode'] = resultCode
caseFrame['dressCode'] = dressCode
    
```

```python
courtFrame = toWrite[toWrite['resultCode'] != "U"].groupby(['court','resultCode']).count()
courtFrame = courtFrame.drop(columns=['title','tid','date','judge','dressMentioned', 'dressCode'])
courtFrame = courtFrame.drop(labels=['Andhra Pradesh High Court - Amravati', 'Jammu & Kashmir High Court - Srinagar Bench', 'Patna High Court - Orders'], axis=0)

pd.set_option("display.max_rows", None)
courtFrame 
```

```python
convictionRates = {} # A:C
count = 0
rate = 0
for index, row in courtFrame.iterrows():
    #print(index)
    if (count % 2 == 0):
        rate += row['resolution'] # adding acquittals
    elif (count % 2 == 1):
        rate /= row['resolution'] # dividing by convitions
        convictionRates[index[0]] = rate #storing
        rate = 0 #resetting
    count += 1
    
convictionFrame = pd.DataFrame.from_dict(convictionRates, orient='index')
avgConviction = convictionFrame[0].mean()
convictionFrame.head(25)
```

```python
judgeFrame = toWrite.copy(deep=True)
judgeList = []
for index, row in judgeFrame.iterrows():
    judgeString = row['judge']
    if(judgeString.lower() == 'unknown'):
        judgeList.append([])
    else:
        judgeParts = judgeString.split(":")
        if(judgeParts[0] == "Bench" or judgeParts[0] == "Author"):
            judges_str = judgeParts[1][1:]
            judges = judges_str.split(", ")
            judgeList.append(judges)
        else:
            judgeList.append([])
        
judgeFrame['judgeList'] = judgeList
judgeFrame.head()
```

```python
judgeDict = {}
for index, row in judgeFrame.iterrows():
    judgeList = row['judgeList']  
    if(row['resultCode'] == "U"):
        continue
    elif(len(judgeList)>0):
        for judge in judgeList:
            #print(judge)
            if judge in judgeDict.keys():
                judge_entry = judgeDict[judge]
                if(row['resultCode']) == 'C':
                    judge_entry['convictions'] += 1
                    if(row['dressCode'] == 'Y'):
                        judge_entry['conviction_clothes'] += 1
                elif(row['resultCode'] == 'A'):
                    judge_entry['acquittals'] += 1
                    if(row['dressCode'] == 'Y'):
                        judge_entry['acquittal_clothes'] += 1
            else:
                judge_entry = {'convictions': 0, 'acquittals': 0, 'conviction_clothes': 0, 'acquittal_clothes': 0}
                if(row['resultCode']) == 'C':
                    judge_entry['convictions'] += 1
                    if(row['dressCode'] == 'Y'):
                        judge_entry['conviction_clothes'] += 1
                elif(row['resultCode'] == 'A'):
                    judge_entry['acquittals'] += 1
                    if(row['dressCode'] == 'Y'):
                        judge_entry['acquittal_clothes'] += 1
                judgeDict[judge] = judge_entry

judgeResults = pd.DataFrame.from_dict(judgeDict, orient='index')
judgeResults
```

```python
majorJudges = judgeResults[(judgeResults['acquittals'] + judgeResults['convictions'] > 3)]
majorJudges[(majorJudges['acquittals'] / majorJudges['convictions'] > 2)]
```
