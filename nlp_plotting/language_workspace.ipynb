```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

casetexts = caseFrame = pd.read_json('../caseData_all.json')
textlist = casetexts['full_text'].to_list()
alltexts = "\n".join(textlist)
```

```python
alltexts[:1000]
```

```python
# importing all necessary modules
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
 
warnings.filterwarnings(action = 'ignore')
 
import gensim
from gensim.models import Word2Vec
 
f = alltexts.replace("\n", " ")
data = []
print("Tokenizing Sentences...")
sentence_tokens = sent_tokenize(f)
# iterate through each sentence in the file
for i, tok in enumerate(sentence_tokens):
    temp = []
    if(i%100 == 0):
        print(i,"of",len(sentence_tokens), 'complete') 
    # tokenize the sentence into words
    for j in word_tokenize(tok):
        temp.append(j.lower())
    data.append(temp)
    
# Create Common Bag Of Words model
print("All tokens generated! Building Bag of Words model...")
model1 = gensim.models.Word2Vec(data, min_count = 1,
                              vector_size = 100, window = 5)
 
# Print results
print("Cosine similarity between 'clothes' " +
               "and 'sexual assault' - CBOW : ",
    model1.wv.similarity('clothes', 'sexual assault'))
     
print("Cosine similarity between 'dress' " +
                 "and 'sexual assault' - CBOW : ",
      model1.wv.similarity('dress', 'sexual assault'))

print("Building Skip Gram model...")
# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count = 1, vector_size = 100,
                                             window = 5, sg = 1)
 
# Print results
print("Cosine similarity between 'clothes' " +
          "and 'sexual assault' - Skip Gram : ",
    model2.wv.similarity('clothes', 'sexual assault'))
     
print("Cosine similarity between 'alice' " +
            "and 'machines' - Skip Gram : ",
      model2.wv.similarity('dress', 'sexual assault'))
```
